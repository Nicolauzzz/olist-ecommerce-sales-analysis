```python
missing_summary = []

for name, df in datasets.items():

    missing = df.isnull().sum()

    percent = (missing/len(df)*100).round(2)

    for col in df.columns:

        missing_summary.append({

            "Dataset": name,

            "Column": col,

            "Missing": missing[col],

            "Percentage": percent[col]

        })

missing_summary = pd.DataFrame(missing_summary)

missing_summary.sort_values(
    "Missing",
    ascending=False
)
```

```python
duplicates = []

for name, df in datasets.items():

    duplicates.append({

        "Dataset": name,

        "Duplicate Rows": df.duplicated().sum()

    })

duplicates = pd.DataFrame(duplicates)

duplicates
```

# Initial Findings

## 1. Dataset Overview

The Olist Brazilian E-commerce dataset consists of **9 relational tables** containing information about customers, orders, products, sellers, payments, reviews, geolocation, and product category translations.

| Dataset | Rows | Columns |
|---------|------:|--------:|
| Customers | 99,441 | 5 |
| Orders | 99,441 | 8 |
| Order Items | 112,650 | 7 |
| Payments | 103,886 | 5 |
| Products | 32,951 | 9 |
| Reviews | 99,224 | 7 |
| Sellers | 3,095 | 4 |
| Geolocation | 1,000,163 | 5 |
| Category Translation | 71 | 2 |

The **Orders** table acts as the central dataset because it connects customers, products, payments, sellers, and reviews.

## 2. Data Quality Assessment

Based on the initial inspection, several data quality issues were identified.

### Missing Values

The following columns contain missing values:

| Dataset | Column | Missing |
|---------|---------|--------:|
| Reviews | review_comment_title | 87,656 |
| Reviews | review_comment_message | 58,247 |
| Orders | order_delivered_customer_date | 2,965 |
| Orders | order_delivered_carrier_date | 1,783 |
| Products | Product attributes | 610 |

Most missing values occur in the review comment columns because not every customer leaves textual feedback after completing an order.

The missing delivery timestamps are likely associated with orders that were canceled or had not yet been delivered.

Product-related missing values represent only a small percentage (approximately **1.85%**) of the dataset.

## 3. Duplicate Records

Duplicate record inspection shows that almost every dataset does not contain duplicated rows.

However, the **Geolocation** dataset contains **261,831 duplicated records**.

These duplicates are expected because multiple customers may share the same ZIP code and geographic coordinates.

The necessity of removing these duplicates will be evaluated during the data cleaning stage.

## 4. Data Structure

The datasets follow a relational database structure.

Main relationships include:

- Customers → Orders
- Orders → Order Items
- Order Items → Products
- Order Items → Sellers
- Orders → Payments
- Orders → Reviews
- Products → Category Translation

This structure allows multiple business analyses such as sales performance, customer behavior, product analysis, seller performance, and logistics.

## 5. Data Readiness

Overall, the dataset is well structured and suitable for analysis.

Before conducting exploratory data analysis, several preprocessing tasks are required:

- Convert timestamp columns into datetime format.
- Handle missing values.
- Remove unnecessary duplicate records (if applicable).
- Merge related tables into a unified analytical dataset.
