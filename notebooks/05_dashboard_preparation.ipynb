# Dashboard Data Preparation

This notebook prepares the final dataset for Power BI.

The objectives are:

- Create a clean dashboard-ready dataset
- Validate data quality
- Export the final dataset
- Prepare KPI metrics

```python
import pandas as pd
```

```python
sales = pd.read_csv("../data/processed/sales_clean.csv")

sales.head()
```

```python
dashboard = sales[
    [
        "order_id",
        "customer_unique_id",
        "customer_city",
        "customer_state",
        "purchase_year",
        "purchase_month",
        "purchase_day",
        "product_category_name_english",
        "price",
        "freight_value",
        "total_sales",
        "payment_type",
        "payment_installments",
        "review_score",
        "delivery_days",
        "order_status"
    ]
]
```

```python
dashboard.head()
```

```python
dashboard = dashboard.rename(
    columns={
        "customer_unique_id":"Customer ID",
        "customer_city":"City",
        "customer_state":"State",
        "purchase_year":"Year",
        "purchase_month":"Month",
        "purchase_day":"Day",
        "product_category_name_english":"Category",
        "price":"Product Price",
        "freight_value":"Shipping Cost",
        "total_sales":"Revenue",
        "payment_type":"Payment Method",
        "payment_installments":"Installments",
        "review_score":"Review Score",
        "delivery_days":"Delivery Days",
        "order_status":"Order Status"
    }
)

dashboard.head()
```

```python
repeat_rate = (
    dashboard
    .groupby("Customer ID")["order_id"]
    .nunique()
)

(
    (repeat_rate>1).sum()
    /
    repeat_rate.count()
    *100
)
```

```python
dashboard["Date"] = pd.to_datetime(
    sales["order_purchase_timestamp"]
)
```

```python
dashboard = dashboard.sort_values("Date")
```

```python
dashboard.head()
```

```python
dashboard.tail()
```

```python
dashboard.to_csv(
    "../data/processed/dashboard_data.csv",
    index=False
)
```
